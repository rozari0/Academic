```python
def bubble_sort(array):
    i = 0
    length = len(array)

    while i < length - 1:
        is_swapped = False
        j = 0
        while j < length - i - 1:
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
                is_swapped = True
            j += 1

        if not is_swapped:
            break

        i += 1

    return array


# array = list(map(int, input("Input numbers: ").split()))
# print(bubble_sort(array))
```

```python
def insertion_sort(array):
    i = 1
    length = len(array)
    while i < length:
        key = array[i]
        j = i - 1

        while j >= 0 and array[j] > key:
            array[j + 1] = array[j]
            j -= 1

        array[j + 1] = key
        i += 1

    return array
```

```python
def selection_sort(array):
    i = 0
    length = len(array)

    while i < length - 1:
        minimum = i
        j = i + 1
        while j < length:
            if array[minimum] > array[j]:
                minimum = j
            j += 1
        array[i], array[minimum] = array[minimum], array[i]
        i += 1

    return array
```

```python
print(bubble_sort([9, 8, 7, 5, 6, 4, 3, 2, 1]))
print(insertion_sort([9, 8, 7, 5, 6, 4, 3, 2, 1]))
print(selection_sort([9, 8, 7, 5, 6, 4, 3, 2, 1]))
```

```python
def selection_sort_trace(array):
    length = len(array)
    i = 0

    while i < length - 1:
        minimum = i
        j = i + 1
        print(f"Pass {i + 1} (i = {i}):")
        while j < length:
            print(
                f"  Compare A[{minimum}] = {array[minimum]} and A[{j}] = {array[j]}",
                end="",
            )
            if array[minimum] > array[j]:
                minimum = j
                print(" → new minimum")
            else:
                print(" → no change")
            j += 1
        if minimum != i:
            array[i], array[minimum] = array[minimum], array[i]
            print(f"  Swap A[{i}] and A[{minimum}] → {array}")
        else:
            print(f"  No swap needed → {array}")
        i += 1
        print()

    print("Sorted array:", array)
    return array


# Example usage
arr = [9, 8, 3, 1, 7]
selection_sort_trace(arr)
```
