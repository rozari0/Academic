```python
from time import perf_counter
```

```python
def bubble_sort(array):
    i = 0
    length = len(array)

    comparison, swap = 0, 0
    start_time = perf_counter()
    while i < length - 1:
        is_swapped = False
        j = 0
        while j < length - i - 1:
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
                is_swapped = True
                swap += 1
            j += 1
            comparison += 1

        if not is_swapped:
            break

        i += 1
    end_time = perf_counter()

    return array, comparison, swap, end_time - start_time
```

```python
def insertion_sort(array):
    i = 1
    length = len(array)

    comparison, swap = 0, 0
    start_time = perf_counter()
    while i < length:
        key = array[i]
        j = i - 1

        while j >= 0:
            comparison += 1
            if array[j] > key:
                array[j + 1] = array[j]
                swap += 1
            else:
                break
            j -= 1

        array[j + 1] = key
        i += 1
    end_time = perf_counter()
    return array, comparison, swap, end_time - start_time
```

```python
def selection_sort(array):
    i = 0
    length = len(array)

    comparison, swap = 0, 0
    start_time = perf_counter()
    while i < length - 1:
        minimum = i
        j = i + 1
        while j < length:
            comparison += 1
            if array[minimum] > array[j]:
                minimum = j
            j += 1

        if minimum != i:
            array[i], array[minimum] = array[minimum], array[i]
            swap += 1
        i += 1
    end_time = perf_counter()
    return array, comparison, swap, end_time - start_time
```

```python
print(bubble_sort([9, 8, 7, 5, 6, 4, 3, 2, 1]))
print(insertion_sort([9, 8, 7, 5, 6, 4, 3, 2, 1]))
print(selection_sort([9, 8, 7, 5, 6, 4, 3, 2, 1]))
```
