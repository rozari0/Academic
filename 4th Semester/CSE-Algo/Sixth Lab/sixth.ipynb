```python
# 1. Maximum subarray sum
array = [1, 2, 3, 4, 5, -6, 8, 9]
max_sum = array[0]
i = 0
while i < len(array):
    j = i
    sum = 0
    while j < len(array):
        sum += array[j]
        max_sum = max(sum, max_sum)
        j += 1
    i += 1

print(max_sum)
```

```python
# 1.1 Kadandes algorithm
array = [-9, -7, -3, -5, -3]
max_sum = array[0]
i = 0
while i < len(array):
    max_sum += array[i]
    if (max_sum + array[i]) < 0:
        max_sum = 0

    i += 1

print(max_sum)
```

```python
# Only Pluses
a, b, c = list(map(int, input("Input numbers: ").split(" ")))
for _ in range(5):
    if a <= b and a <= c:
        a += 1
    elif b <= a and b <= c:
        b += 1
    else:
        c += 1

print(a * b * c)
```

```python
# 3. Trailing Zero in Factorial

num: int = int(input("Input number: "))
count = 0
i = 0
while i <= num:
    count += i // 5
    i += 5

print(int(count))
```

```python
# 4. Digit of Factorial
from math import log, floor

num: int = int(input("Input a number: "))
count: int = 1

for i in range(1, num + 1):
    count += log(i, 10)


print(floor(count))
```
