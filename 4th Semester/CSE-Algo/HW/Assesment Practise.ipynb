```python
from typing import List
from random import randint
```

```python
## Linear Search
def linear_search(array: List[int], search_item):
    index: int = -1
    for i, item in enumerate(array):
        if item == search_item:
            index = i
            break
    return index


array = input("Input array: ")
search_item = int(input("Input search item: "))
array = list(map(int, array.split()))

print(linear_search(array, search_item))
```

```python
## Binary search


def binary_search(array: List[int], search_item: int) -> int:
    length = len(array)
    left, right = 0, length - 1
    mid = (left + right) // 2

    while left <= right:
        if array[mid] == search_item:
            return mid
        elif array[mid] < search_item:
            left = mid + 1
        else:
            right = mid - 1

        mid = (left + right) // 2

    return -1


index = binary_search([1, 3, 4, 9, 15], 8)
print(index)
```

```python
# Bubble sort


def bubble_sort(array: List[int]):
    length = len(array)

    for i in range(length):
        for j in range(length - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]

    return array


array = [randint(1, 100) for _ in range(20)]
bubble_sort(array)
```

```python
## Selection sort
def selection_sort(array: List[int]):
    length = len(array)
    for i in range(length):
        key = i
        for j in range(i, length):
            if array[key] > array[j]:
                key = j
        array[key], array[i] = array[i], array[key]
        print(array)


selection_sort([7, 12, 9, 11, 3])
```

```python
## Insertion sort
def insertion_sort(array: List[int]):
    length = len(array)

    for i in range(1, length):
        key = array[i]
        j = i - 1

        while j >= 0 and array[j] > key:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key
    return array


insertion_sort([7, 12, 9, 11, 3])
```

```python
# Kadan's Algorithm


def kadanes_algorithm(array: List[int]):
    max_sum = float("-inf")
    curr_sum = 0

    for item in array:
        curr_sum = max(curr_sum, 0)
        curr_sum += item

        max_sum = max(curr_sum, max_sum)
    return max_sum


kadanes_algorithm([-1, -2, -3, -4, 0])
```
