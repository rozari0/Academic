```python
from typing import List


class OnlineSorting:
    array = None

    def __init__(self, array: None | List[int] = None):
        if array:
            self.array = array
        else:
            self.array = []

    def add(self, key: int):
        j = len(self.array) - 1
        while j >= 0 and self.array[j] > key:
            j -= 1
        self.array.insert(j + 1, key)
        print(f"After inserting {key}: {self.array}")
```

```python
array = OnlineSorting()

limit: int = int(input())
while limit:
    array.add(int(input()))
    limit -= 1
```

```python
array = OnlineSorting()

while True:
    number = int(input("Input a number: "))
    array.add(number)
    choice = input("Do you want to insert another number? (y/n): ")
    if choice.lower() != "y":
        print(f"Complete array {array.array}")
        break
```
