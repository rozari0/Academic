```python
def longest_common_sub(st1: str, st2: str):
    len1, len2 = len(st1), len(st2)

    mat = [[0] * (len2 + 1) for _ in range(len1 + 1)]

    for i in range(1, len1 + 1):
        for j in range(1, len2 + 1):
            if st1[i - 1] == st2[j - 1]:
                mat[i][j] = mat[i - 1][j - 1] + 1
            else:
                mat[i][j] = max(mat[i - 1][j], mat[i][j - 1])

    i, j = len1, len2

    lcs = []
    while i and j:
        if st1[i - 1] == st2[j - 1]:
            lcs.append(st1[i - 1])
            i, j = i - 1, j - 1
        elif mat[i - 1][j] > mat[i][j - 1]:
            i -= 1
        else:
            j -= 1

    return mat[len1][len2], "".join(reversed(lcs))


longest_common_sub("greedy", "forgive")
```
